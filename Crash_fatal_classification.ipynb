```python
X = df.drop(columns=["IS_FATAL"])
y = df["IS_FATAL"]
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix

dt = DecisionTreeClassifier(
    max_depth=None,
    min_samples_split=2,
    random_state=42
)

dt.fit(X_train_scaled, y_train)
dt_pred = dt.predict(X_test_scaled)

print("----- Decision Tree Results -----")
print(confusion_matrix(y_test, dt_pred))
print(classification_report(y_test, dt_pred))
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
import numpy as np


dt_params = {
    "model__max_depth": [None, 5, 10, 15, 20],
    "model__min_samples_split": [2, 5, 10],
    "model__min_samples_leaf": [1, 2, 4],
    "model__criterion": ["gini", "entropy"]
}

dt_search = RandomizedSearchCV(
    dt_pipe,
    dt_params,
    n_iter=10,          
    cv=3,                 
    scoring="f1",
    n_jobs=-1,
    random_state=42
)

dt_search.fit(X_train, y_train)

print("Best Decision Tree Params:", dt_search.best_params_)
best_dt = dt_search.best_estimator_
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

models = {
    "Decision Tree": dt,            
    "Tuned Decision Tree": best_dt, 
    "Random Forest": rf,            
    "Fast Random Forest": fast_rf   
}

for name, model in models.items():
    
   
    preds = model.predict(X_test)
    
    cm = confusion_matrix(y_test, preds)
    
    # Plot
    plt.figure(figsize=(5,4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(f"{name} - Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.tight_layout()
    plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd

results = []

for name, model in models.items():   
    preds = model.predict(X_test)
    
    results.append({
        "Model": name,
        "Accuracy": accuracy_score(y_test, preds),
        "Precision": precision_score(y_test, preds),
        "Recall": recall_score(y_test, preds),
        "F1 Score": f1_score(y_test, preds)
    })

results_df = pd.DataFrame(results)
print("----- Model Comparison Table -----")
print(results_df)
```

```python
best_model = results_df.loc[results_df['F1 Score'].idxmax()]
print("\nBest Model Based on F1 Score:")
print(best_model)
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

plt.figure(figsize=(7,6))

for name, model in models.items():

    # only models with predict_proba can do ROC
    if hasattr(model, "predict_proba"):
        y_prob = model.predict_proba(X_test)[:, 1]
        fpr, tpr, _ = roc_curve(y_test, y_prob)
        roc_auc = auc(fpr, tpr)

        plt.plot(fpr, tpr, label=f"{name} (AUC = {roc_auc:.3f})")

plt.plot([0,1], [0,1], 'k--')
plt.title("ROC Curve Comparison")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np

def bootstrap_once(X, y, random_state=None):
    rng = np.random.default_rng(random_state)
    n = len(X)
    boot_idx = rng.integers(0, n, n)

    X_boot = X.iloc[boot_idx]
    y_boot = y.iloc[boot_idx]

    oob_idx = np.setdiff1d(np.arange(n), boot_idx)
    X_oob = X.iloc[oob_idx]
    y_oob = y.iloc[oob_idx]

    return X_boot, y_boot, X_oob, y_oob
```

```python
from sklearn.tree import DecisionTreeClassifier

bootstrap_models = []
n_models = 10   

for i in range(n_models):
    Xb, yb, X_oob, y_oob = bootstrap_once(X_train, y_train, random_state=i)

    model = DecisionTreeClassifier(random_state=i)
    model.fit(Xb, yb)

    bootstrap_models.append((model, X_oob, y_oob))
```

```python
predictions = []

for model, _, _ in bootstrap_models:
    predictions.append(model.predict(X_test))

pred_array = np.array(predictions)
final_pred = (pred_array.mean(axis=0) > 0.5).astype(int)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print("----- Bootstrap Ensemble Confusion Matrix -----")
print(confusion_matrix(y_test, final_pred))

print("----- Bootstrap Ensemble Classification Report -----")
print(classification_report(y_test, final_pred))
```
